```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import pandas as pd

%matplotlib inline
data = pd.read_csv("../data/processed/train.csv")
```

```python
pos = np.where(data.iloc[:, 0] == 1)[0]
neg = np.where(data.iloc[:, 0] == 0)[0]

plt.figure(figsize=(8,10))
plt.bar([1, 2], [len(pos), len(neg)], color=['b', 'r'])
plt.xticks([1, 2], ['Positive', 'Negative'])
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.2%}'.format(float(y) / data.shape[0])))

plt.show()
```
